```python
import agate
```

```python
lust1CLEANED = agate.Table.from_csv('/Users/reeceristau/Downloads/lust1CLEANEDutf8.csv')
```

I imported my cleaned data.

```python
print(lust1CLEANED)
```

```python
lust1 = lust1CLEANED.where(lambda row: row['S'] != '5')
print(len(lust1.rows))
```

```python
lust2 = lust1.where(lambda row: row['S'] != '6')
print(len(lust2.rows))
```

```python
lust3 = lust2.where(lambda row: row['S'] != '7')
print(len(lust3.rows))
```

```python
lust4 = lust3.where(lambda row: row['S'] != '8')
print(len(lust4.rows))
```

```python
lust5 = lust4.where(lambda row: row['S'] != 'R')
print(len(lust5.rows))
```

```python
lust6 = lust5.where(lambda row: row['S'] != 'V')
print(len(lust6.rows))
```

The above 6 steps filtered out the stuff we didn't want.

```python
groups = lust6.group_by('OWNCO--------------------')
```

```python
groups_count = groups.aggregate([
        ('count', agate.Length())
    ])
```

This grouped the data by OWNCO and counted it.

```python
groups_count.print_table()
```

```python
sorted_groups = groups_count.order_by('count', reverse=True)
sorted_groups.print_table(max_rows=20)
```

Finally, I sorted it by count, reversed the order and printed it with 20.
