```python
import agate
```

```python
frl13 = agate.Table.from_csv('/Users/reeceristau/GitHub/JOUR407-Data-Journalism/Data/frl13.csv')
frl14 = agate.Table.from_csv('/Users/reeceristau/GitHub/JOUR407-Data-Journalism/Data/frl14.csv')
frl15 = agate.Table.from_csv('/Users/reeceristau/GitHub/JOUR407-Data-Journalism/Data/frl15.csv')
print(frl13)
```

This got all three files in agate.

```python
frl = frl13.join(frl14, 'CODISTSCH13', 'CODISTSCH14', inner=True)
```

```python
frl = frl.join(frl15, 'CODISTSCH13', 'CODISTSCH15', inner=True)
```

The above two steps joined '13 and '14 and then that table with '15

```python
print(frl)
print(len(frl.rows))
```

```python
new_table=frl.select(['COUNTY13', 'CODISTSCH13', 'FREEREDUCED13', 'FREEREDUCED14', 'FREEREDUCED15'])
```

This selected the columns I wanted.

```python
print(new_table)
print(len(new_table.rows))
```

```python
new_table = new_table.where(lambda row: row['FREEREDUCED13'] != None)
print(len(new_table.rows))
```

```python
new_table = new_table.where(lambda row: row['FREEREDUCED15'] != None)
print(len(new_table.rows))
```

The above two bullshit steps got rid of that blank column.

```python
change = new_table.compute([
    ('percent_change', agate.PercentChange('FREEREDUCED13', 'FREEREDUCED15'))       
])
```

```python
change = change.order_by('percent_change', reverse=True)
```

```python
for_printing = change.select(['COUNTY13', 'percent_change'])
```

```python
for_printing.print_table()
```

Finally, I ordered it by percent change and printed it.
