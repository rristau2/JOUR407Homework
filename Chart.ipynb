```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

sns.set(style="ticks")
%matplotlib inline
```

```python
ml = pd.read_csv('mountainlions.csv')
```

```python
print(ml)
```

```python
sns.set_style("whitegrid")
sns.set_color_codes("dark")
sns.barplot(x="count", y="COUNTY", data=ml, color="b")
```

```python
sns.set_style("ticks")
sns.set_color_codes("dark")
sns.barplot(x="count", y="COUNTY", data=ml, color="r")
```

```python
sns.set_style("darkgrid")
sns.set_color_codes("dark")
sns.barplot(x="count", y="COUNTY", data=ml, color="y")
```
